```python
from random import seed
from random import randint
from numpy import array
from sklearn.utils import validation
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from math import sqrt
from sklearn.metrics import mean_squared_error
import random
from random import seed
import matplotlib.pyplot as pyplot
import numpy as np

def multiplyList(myList):
    result = 1
    for x in myList:
        result *= x
    return result

def random_sum_pairs(n_examples, n_numbers, largest):
    X, y = list(), list()
    for i in range(n_examples):
        in_pattern = [round(random.uniform(1.0, 100.0),2) for _ in range(n_numbers)]  # 2 değer ataması için
        out_pattern = round(multiplyList(in_pattern),2)
        X.append(in_pattern)
        y.append(out_pattern)
        
    # Numpy Formatına Çevrildi
    X, y = array(X), array(y)
    # Sayılar Normalize Edildi 
    X = X.astype('float') / float(largest * n_numbers)
    y = y.astype('float') / float(largest * n_numbers)
    return X, y
```

```python


# Normalizasyonu Tersine çevirme Fonksiyonu
def invert(value, n_numbers, largest):
    return value * float(largest * n_numbers)

seed(1)
n_examples = 100
n_numbers = 2
largest = 100.0
# LSTM konfigürasyonunu tanımlandı
n_batch = 1
n_epoch = 100
model = Sequential()
model.add(LSTM(10, input_shape=(n_numbers, 1)))
model.add(Dense(1))
model.compile(loss='MSE', optimizer='adam')
# LSTM 
for __ in range(n_epoch):
    X, y = random_sum_pairs(n_examples, n_numbers, largest)
    X = X.reshape(n_examples, n_numbers, 1)
    history = model.fit(X, y, epochs=20, batch_size=n_batch, verbose=2)
    print('=====================================================')
# Eğitim Ve Loss Grafiğinin Çizdirilmesi 
pyplot.plot(history.history['loss'])
pyplot.title('model train')
pyplot.ylabel('loss')
pyplot.xlabel('epoch')
pyplot.legend(['train', 'loss'], loc='upper right')
pyplot.show()


X, y = random_sum_pairs(n_examples, n_numbers, largest)
X = X.reshape(n_examples, n_numbers, 1)
result = model.predict(X, batch_size=n_batch, verbose=0)
# Hatanın Hesaplanması 
values = [invert(aa, n_numbers, largest) for aa in X]
expected = [invert(x, n_numbers, largest) for x in y]
predicted = [invert(x, n_numbers, largest) for x in result[:, 0]]  #
rmse = sqrt(mean_squared_error(expected, predicted)) 
print('RMSE: %f' % rmse)  # RMSE karesel hata
# Örneklerin Gösterilmesi 
for i in range(10):
    error = expected[i] - predicted[i]
    print("Inputs = ",values[i][0], "*", values[i][1],"Expected =",round(expected[i],2), "Predicted =", round(predicted[i],2),"err =",round(error,2))
```
